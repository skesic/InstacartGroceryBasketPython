## 13. Explore whether there's a connection between age and spending power (income). Create a scatterplot using sns.scatterplot() function. In a markdown cell beneath scatterplot, describe what the scatterplot showes.

#### From this graph we can observe a connection between age and income. We can see that the the vast majority, in all age groups, tend to earn under 200,000. Around age 40+ more of population is earning up to 300,000. Between 20's and 40's no one is earning more than 400,000. We can also observe that between 40's and 80's people tend to earn no more than 600,000. Our conclusion is that income generally tends to increase after 40 years of age, which makes sense, as people at that time get promotions and work on senior levels, and it is normal increase in salery for that life period.  We can conclude that people mostly earn till 40 years of age around 200,000, although there exist some percentage that earn more than that, but at the same time less than 400,000. After age of 40 majority earns around 300,000 although some can earn up to 600,000.

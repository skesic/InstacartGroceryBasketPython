#### 05. Creating a Function

```python
# Creating a function

def price_label(row):
    
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range products'
    else: return np.nan
```

#### 06. Applying the Function

```python
# Apply the function

df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Count the number of products within each label

df['price_range'].value_counts(dropna = False)
```

#### 08. If - Statements with the loc() Function

```python
# If the item's price is above $15, it will be labeled a "High-range product"

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
# If the item's price is above $5 but lower than or equal to $15, it will be labeled a "mid-range product"

df.loc[(df['prices'] <= 15) &(df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
# If the item's price is lower than or equal to $5, it will be labeled a "Low-range product"

df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# Count the number of products within each label

df['price_range_loc'].value_counts(dropna = False)
```

### 04. Identify the busiest hours of the day. Rather than by hour, periods of time should be labeled "Most orders", "Average orders", and "Fewest orders". Create a new column containing these labels called "busiest_period_of_day".

```python
# Check df_ords_prods_merged

df_ords_prods_merged.head()
```

```python
# On which hour most orders take place

df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

#### Creating a new column "busiest_period_of_day"

```python
# Creating a column "busiest period of day"

result3 = []

for value in df_ords_prods_merged["order_hour_of_day"]:
    if value == 10 or value == 11 or value == 14 or value == 15 or value == 13 or value == 12 or value == 16 or value ==9:
        result3.append("Most orders")
    elif value == 23 or value == 6 or value == 0 or value == 1 or value == 5 or value == 2 or value == 4 or value == 3:
        result3.append("Fewest orders")
    else:
        result3.append("Average orders")
```

```python
# Running a loop

result3
```
